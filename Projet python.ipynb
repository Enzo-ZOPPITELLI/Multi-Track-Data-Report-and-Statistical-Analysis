### Uploader la data

```python
# Upload du fichier excel sur python

excel_file = 'student_grades_2027-2028.xlsx'

all_sheets = pd.read_excel(excel_file, sheet_name=None, engine='openpyxl')
all_sheets
```

```python
# Affiche le nom des feuilles dans le fichier excel student_grades_2027-2028.xlsx
print(f"Feuilles trouvées dans le fichier : {list(all_sheets.keys())}") 
```

```python
# Création d'une boucle pour afficher le nom et le nombre de lignes ayant des valeurs de chaque feuille

frames = []

for sheet_name, df in all_sheets.items():
    print(f"\nChargement de la feuille : {sheet_name} ({len(df)} lignes)")
    df = df.copy()

    # On ajoute une colonne pour savoir de quelle filière vient chaque ligne
    df["Track"] = sheet_name

    # Nettoyer les valeurs spéciales comme "N/A", "Waived", etc.
    df.replace(["N/A", "Waived", "-", "--", "None", "null"], np.nan, inplace=True)

    # Ajouter la feuille nettoyée à la liste
    frames.append(df)
```

```python
# Fusion en un même fichier

merged_df = pd.concat(frames, ignore_index=True)

print(f"\nDonnées fusionnées : {merged_df.shape[0]} lignes × {merged_df.shape[1]} colonnes")
```

```python
# Cherche les colonnes contenant des chiffres, enlève les % et 
# espaces, puis convertit ces colonnes en valeurs numériques quand c’est possible.

for col in merged_df.columns:
    if merged_df[col].dtype == object:  # texte
        # Si la colonne ressemble à des notes ou des pourcentages
        if merged_df[col].astype(str).str.contains(r"\d").any():
            merged_df[col] = merged_df[col].astype(str).str.replace("%", "", regex=False).str.strip()
            merged_df[col] = pd.to_numeric(merged_df[col], errors="ignore")
```

```python
# Crée le fichier excel avec toutes les colonnes modifiées + vérification de la création

output_file = "merged_clean_part1.csv"
merged_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f" Données nettoyées enregistrées dans : {output_file}")
```

## Cross-Track Comparative Analysis

```python
import matplotlib.pyplot as plt
```

```python
print("Colonnes disponibles :", df.columns.tolist) 
# Connaître le nombre de colonnes disponibles / Fonction de chaque colonnes
```

```python
# On reconverti les colonnes à la bonne forme

def clean_numeric(s):
    return (
        s.astype(str)
        .str.replace("%", "", regex=False)
        .str.strip()
        .replace(["N/A", "Waived", "None", "-", "--"], None)
        .astype(float)
    )

for col in ["Mathematics", "History", "ProjectScore", "Attendance (%)"]:
    if col in df.columns:
        df[col] = clean_numeric(df[col])
```

```python
# Visualisation de la distribution des notes d'histoires sous la forme d'un histogram

if "History" in df.columns:
    for track, subset in df.groupby("Track"):
        plt.figure()
        subset["History"].dropna().plot(
            kind="hist",
            bins=10,
            color="red",
            title=f"Distribution des notes d'Histoire — {track}"
        )
        plt.xlabel("Note d'Histoire")
        plt.ylabel("Nombre d'étudiants")
        plt.tight_layout()
        plt.savefig(f"hist_history_{track}.png")
        plt.close()

    # Boxplot global
    plt.figure()
    data = [subset["History"].dropna() for _, subset in df.groupby("Track")]
    labels = [track for track, _ in df.groupby("Track")]
    plt.boxplot(data, labels=labels)
    plt.title("Comparaison des notes d'Histoire par filière")
    plt.ylabel("Note d'Histoire")
    plt.tight_layout()
    plt.savefig("boxplot_history_by_track.png")
    plt.close()
```

```python
# Moyenne des notes de mathématiques et en fait un bar-chart

if "Mathematics" in df.columns:
    math_means = df.groupby("Track")["Mathematics"].mean().sort_values()
    plt.figure()
    math_means.plot(kind="bar", title="Moyenne en Mathématiques par filière")
    plt.ylabel("Moyenne en Mathématiques")
    plt.tight_layout()
    plt.savefig("bar_math_mean_by_track.png")
    plt.close()
```

```python
# pour chaque filière, la corrélation entre l’assiduité et 
# le score de projet, puis génère un scatter plot 

att_col = "Attendance (%)" if "Attendance (%)" in df.columns else None
proj_col = "ProjectScore" if "ProjectScore" in df.columns else None

if att_col and proj_col:
    corr_data = []
    for track, subset in df.groupby("Track"):
        subset = subset[[att_col, proj_col]].dropna()
        if len(subset) > 1:
            corr = subset[att_col].corr(subset[proj_col])
            corr_data.append({"Track": track, "Corr_Attendance_Project": round(corr, 3)})
            # Scatter plot
            plt.figure()
            plt.scatter(subset[att_col], subset[proj_col])
            plt.title(f"Présence vs Score de projet — {track}")
            plt.xlabel(att_col)
            plt.ylabel(proj_col)
            plt.tight_layout()
            plt.savefig(f"scatter_attendance_project_{track}.png")
            plt.close()

# Enregistre les résultats dans un fichier CSV.
    
    corr_df = pd.DataFrame(corr_data)
    corr_df.to_csv("attendance_project_correlation.csv", index=False, encoding="utf-8-sig")
    print("Corrélations sauvegardées dans attendance_project_correlation.csv")

print("Graphiques enregistrés (histogrammes, boxplots, barres, corrélations)")
```

## Cohort-Level Analysis

```python
# Cherche le bon nom de colonne pour la cohorte, 
# le statut de revenu et le pass/fail parmi plusieurs possibilités

cohort_col = None
for cand in ["Cohort", "CohortName", "Cohort_ID"]:
    if cand in df.columns:
        cohort_col = cand
        break

income_col = None
for cand in ["IncomeStudent", "Income_Supported", "LowIncome"]:
    if cand in df.columns:
        income_col = cand
        break

pass_col = None
for cand in ["Passed (Y/N)", "Passed", "Pass", "PassYN"]:
    if cand in df.columns:
        pass_col = cand
        break
```

```python
# vérifie la présence des colonnes importantes et avertit (ou stoppe) 
# si l’une d’elles est manquante

if cohort_col is None:
    raise ValueError("Aucune colonne 'Cohort' trouvée dans le fichier.")
if income_col is None:
    print("Aucune colonne 'IncomeStudent' trouvée. Elle sera ignorée.")
if pass_col is None:
    print("Aucune colonne 'Passed' trouvée. Le taux de réussite ne sera pas calculé.")
```

```python
# Retrie la data et converti les données en nombre

score_cols = [c for c in ["Mathematics", "English", "Science", "History", "ProjectScore", "Attendance (%)"]
              if c in df.columns]

def clean_numeric(series):
    return (
        series.astype(str)
        .str.replace("%", "", regex=False)
        .str.strip()
        .replace(["N/A", "Waived", "None", "-", "--"], None)
        .astype(float)
    )

for col in score_cols:
    df[col] = clean_numeric(df[col])
```

```python
# Crée une colonne Pass avec 1 pour YES et 0 pour NO (échec de la validation)

if pass_col:
    df["_pass"] = (
        df[pass_col]
        .astype(str)
        .str.strip()
        .str.upper()
        .map({"Y": 1, "YES": 1, "TRUE": 1, "N": 0, "NO": 0, "FALSE": 0})
    )
```

```python
# convertit la colonne de revenu en True/False, 
# et si aucune colonne n’existe, il en crée une (IncomeStudent) remplie avec False.

if income_col:
    df[income_col] = (
        df[income_col]
        .astype(str)
        .str.strip()
        .str.lower()
        .map({"true": True, "1": True, "yes": True, "y": True, "false": False, "0": False, "no": False, "n": False})
        .fillna(False)
    )
else:
    df["IncomeStudent"] = False
    income_col = "IncomeStudent"
```

```python
# regroupe les données par Cohort et IncomeStudent, puis calcule pour chaque groupe 
# le nombre d’élèves, la moyenne des scores, et éventuellement le taux de réussite, 
# afin de produire un tableau récapitulatif cohort_stats.

group_cols = [cohort_col, income_col]
agg_dict = {col: "mean" for col in score_cols}
agg_dict["StudentID" if "StudentID" in df.columns else "Track"] = "count"
if pass_col:
    agg_dict["_pass"] = "mean"

cohort_stats = (
    df.groupby(group_cols)
      .agg(agg_dict)
      .reset_index()
      .rename(columns={"StudentID": "Number_of_Students", "_pass": "PassRate"})
)
```

```python
cohort_stats["PassRate"] = (cohort_stats["PassRate"] * 100).round(1)
for col in score_cols:
    cohort_stats[col] = cohort_stats[col].round(2)
```

```python
cohort_stats.to_csv("cohort_summary.csv", index=False, encoding="utf-8-sig")

print("Fichier 'cohort_summary.csv' créé avec les colonnes suivantes :")
print(cohort_stats.columns.tolist())
print(cohort_stats.head())
```

## Final Report Generation

```python
import os
```

```python
# On renomme le tout

merged_path = "merged_clean_part1.csv"
track_path  = "track_summary.csv"
cohort_path = "cohort_summary.csv"
corr_path   = "attendance_project_correlation.csv"
```

```python
def ensure_dir(d):
    if not os.path.exists(d):
        os.makedirs(d, exist_ok=True)

def coerce_numeric(series):
    return (
        series.astype(str)
        .str.replace("%", "", regex=False)
        .str.strip()
        .replace({
            "WAIVED": np.nan, "Waived": np.nan, "N/A": np.nan, "NA": np.nan,
            "NaN": np.nan, "None": np.nan, "null": np.nan, "--": np.nan, "-": np.nan, "": np.nan
        })
        .pipe(pd.to_numeric, errors="coerce")
    )

def detect_col(df, candidates, fuzzy_keys=None):
    """Retourne le 1er nom de colonne trouvé parmi candidates. Sinon fuzzy sur fuzzy_keys."""
    for c in candidates:
        if c in df.columns:
            return c
    if fuzzy_keys:
        low = {c.lower(): c for c in df.columns}
        for k in low:
            if any(fk in k for fk in fuzzy_keys):
                return low[k]
    return None
```

```python
df_merged = pd.read_csv(merged_path)  # requis pour le rapport final

# Normaliser quelques colonnes numériques probables
for c in ["Mathematics","English","Science","History","ProjectScore","Attendance (%)","Attendance"]:
    if c in df_merged.columns:
        df_merged[c] = coerce_numeric(df_merged[c])

# TrackSummary : charger si présent, sinon recompute rapide
if os.path.exists(track_path):
    track_stats = pd.read_csv(track_path)
else:
    # Recompute minimal si track_summary absent
    score_cols = [c for c in ["Mathematics","English","Science","History","ProjectScore","Attendance (%)"] if c in df_merged.columns]
    agg = {}
    if "StudentID" in df_merged.columns:
        agg["StudentID"] = "count"
    for c in score_cols:
        agg[c] = "mean"
    if agg:
        track_stats = df_merged.groupby("Track").agg(agg).reset_index()
        if "StudentID" in track_stats.columns:
            track_stats = track_stats.rename(columns={"StudentID":"Number_of_Students"})
        else:
            sizes = df_merged.groupby("Track").size().reset_index(name="Number_of_Students")
            track_stats = sizes.merge(track_stats, on="Track", how="left")
    else:
        track_stats = df_merged.groupby("Track").size().reset_index(name="Number_of_Students")

# PassRate (%) si on peut le (re)calculer proprement
pass_col = detect_col(df_merged, ["Passed (Y/N)","Passed","Pass","PassYN","Result","Status"], fuzzy_keys=["pass","result","status"])
if pass_col:
    pass_map = {"Y":1,"YES":1,"TRUE":1,"T":1,"1":1, "N":0,"NO":0,"FALSE":0,"F":0,"0":0}
    tmp = df_merged[["Track", pass_col]].copy()
    tmp["_pass"] = tmp[pass_col].astype(str).str.strip().str.upper().map(pass_map)
    pr = tmp.groupby("Track")["_pass"].mean().mul(100).round(1).rename("PassRate (%)")
    track_stats["PassRate (%)"] = track_stats["Track"].map(pr)

# CohortSummary : charger si présent, sinon recompute si colonnes dispo
if os.path.exists(cohort_path):
    cohort_stats = pd.read_csv(cohort_path)
else:
    cohort_col = detect_col(df_merged, ["Cohort","CohortName","Cohort_ID"])
    income_col = detect_col(df_merged, ["IncomeStudent","Income_Supported","LowIncome"], fuzzy_keys=["income"])
    if cohort_col is not None:
        if income_col is None:
            income_col = "IncomeStudent"
            df_merged[income_col] = False
        # normaliser income en bool
        df_merged[income_col] = (
            df_merged[income_col].astype(str).str.strip().str.lower()
            .map({"true":True,"1":True,"yes":True,"y":True,"false":False,"0":False,"no":False,"n":False})
            .fillna(False)
        )
        score_cols = [c for c in ["Mathematics","English","Science","History","ProjectScore","Attendance (%)"] if c in df_merged.columns]
        agg = {col:"mean" for col in score_cols}
        if "StudentID" in df_merged.columns:
            agg["StudentID"] = "count"
        else:
            agg["Track"] = "count"
        if pass_col:
            df_merged["_pass"] = df_merged[pass_col].astype(str).str.strip().str.upper().map(pass_map)
            agg["_pass"] = "mean"
        cohort_stats = (
            df_merged.groupby([cohort_col, income_col]).agg(agg).reset_index()
        )
        # renommages/finitions
        if "StudentID" in cohort_stats.columns:
            cohort_stats = cohort_stats.rename(columns={"StudentID":"Number_of_Students"})
        else:
            cohort_stats = cohort_stats.rename(columns={"Track":"Number_of_Students"})
        if "_pass" in cohort_stats.columns:
            cohort_stats["PassRate"] = (cohort_stats["_pass"] * 100).round(1)
            cohort_stats.drop(columns=["_pass"], inplace=True)
        for c in score_cols:
            cohort_stats[c] = cohort_stats[c].round(2)
    else:
        cohort_stats = pd.DataFrame()  # pas de cohorte dispo

# Correlations : charger si présent, sinon recompute rapidement si possible
if os.path.exists(corr_path):
    corr_df = pd.read_csv(corr_path)
else:
    att_col = "Attendance (%)" if "Attendance (%)" in df_merged.columns else ("Attendance" if "Attendance" in df_merged.columns else None)
    proj_col = "ProjectScore" if "ProjectScore" in df_merged.columns else None
    corr_df = pd.DataFrame(columns=["Track","Attendance_Project_Corr"])
    if att_col and proj_col:
        rows = []
        for track, g in df_merged.groupby("Track"):
            d = g[[att_col, proj_col]].dropna()
            if len(d) >= 2:
                r = d[att_col].corr(d[proj_col])
                rows.append({"Track": track, "Attendance_Project_Corr": round(float(r), 3)})
        corr_df = pd.DataFrame(rows)
```

```python
# Création du fichier excel au global

ensure_dir("out")
excel_path = os.path.join("out", "summary_report.xlsx")

with pd.ExcelWriter(excel_path, engine="openpyxl") as writer:
    # Onglet 1 : CleanMerged
    df_merged.to_excel(writer, index=False, sheet_name="CleanMerged")
    # Onglet 2 : TrackSummary
    track_stats.to_excel(writer, index=False, sheet_name="TrackSummary")
    # Onglet 3 : CohortSummary (si non vide)
    if not isinstance(cohort_stats, pd.DataFrame) or cohort_stats.empty:
        # créer tout de même une feuille vide avec juste un message
        pd.DataFrame({"info":["No cohort data available"]}).to_excel(writer, index=False, sheet_name="CohortSummary")
    else:
        cohort_stats.to_excel(writer, index=False, sheet_name="CohortSummary")
    # Onglet 4 : Correlations
    if corr_df.empty:
        pd.DataFrame({"info":["No correlation data available"]}).to_excel(writer, index=False, sheet_name="Correlations")
    else:
        corr_df.to_excel(writer, index=False, sheet_name="Correlations")

print("Rapport Excel créé :", os.path.abspath(excel_path))
```
